**Handling Missing Values**

director, cast, country, and rating have the nulls filled with 'Unavailable'.




```python
df.fillna({'director':'Unavailable', 'cast':'Unavailable', 'country':'Unavilable', 'rating':'Unavailable',}, inplace=True)
df.isna().sum()
```

For nulls in date_added, missing date_added is to be substituted in with the most recent date from date_added. This is because Netflix has the tendency to add more content over time. Other viable options would be finding actual dates and inputting them manually or dropping the data from results since the amount of missing data is rather small.

```python
df[df.date_added.isnull()]
```

```python
most_recent_entry_date=df['date_added'].max()
df.fillna({'date_added':most_recent_entry_date}, inplace=True)
```

proof the concept that the date filled in the null date_added from a show_id example that previously was missing the date_added

```python
df[df.show_id == 's6175']
```

# **Additional Data Cleansing **

**Durations data input error**

The missing durations are all movies by Louis C.K. Normally, we would likely fill the duration with the mean duration movies from the table. In this case it appears the actual duration was input into the rating column, so one solution is to move the rating data into the duration and make the rating information 'Unavailable' like other nulls.


```python
df[df.duration.isnull()]
```

check to make sure there is no other content with the same director to avoid accidental overwritting


```python
df[df.director == 'Louis C.K.'].head()
```

Overwrite and Check

```python
#Loc helps us easily accessing the columns by names
df.loc[df['director']=='Louis C.K.','duration'] = df['rating']
df[df.director=='Louis C.K.'].head()
```

Second overwrite check

```python
df.loc[df['director']=='Louis C.K.','rating'] = 'Unavailable'
df[df.director == 'Louis C.K.'].head()
```

# Visualisations

Let's take a look at type of Shows that has been watched on Netflix

```python
df.type.value_counts() #value_counts method shows the counts of different categories in a given column
```

```python
#countplot helps us to plot counts of each category
sns.countplot(x='type',data=df)
plt.title('Count Vs Type of Shows')
```

On Netflix there are more of movies as compared to Tv shows

# **Country Analysis**

```python
df.country.value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country',order=df['country'].value_counts().index[0:10],data = df)
plt.title('Country Wise Content on Netflix')
```

```python
#now checking type of content based on country
movie_countries = df[df['type']=='Movie']
tvshows_countries = df[df['type'] == 'TV Show']
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country', order = df['country'].value_counts().index[0:10],data=movie_countries)
plt.title('Top 10 countries producing movies in Netflix')

plt.figure(figsize=(12,6))
sns.countplot(y='country',order = df['country'].value_counts().index[0:10], data=tvshows_countries)
plt.title('Top 10 countries producing tv shows in Netflix')
```
